```python
import matplotlib.pyplot as plt
import numpy as np

# Re-defining the data
companies = ["Qualcomm", "Apple", "Samsung", "HiSilicon", "MediaTek", "Google", "NVIDIA"]
values = [106, 44, 16, 10, 9, 2, 1]

# Calculate the cumulative sum for the waterfall effect
cumulative_values = np.cumsum(values)

# Set colors for waterfall (positive and negative)
colors = ['#FF6F61', '#6B5B95', '#88B04B', '#F7CAC9', '#92A8D1', '#D65076', '#45B8AC']

# Start plot
plt.style.use('dark_background')
fig1, ax = plt.subplots(figsize=(10, 6))
fig1.set_facecolor("black")

# Create the bars for the waterfall chart
ax.bar(companies, values, color=colors)

# Add a line showing cumulative values
ax.plot(companies, cumulative_values, color='white', marker='o', linestyle='--', linewidth=2)

# Set chart title and labels
plt.title("Waterfall Chart for Phone Companies by Market Usage", color="white", fontsize=14)
plt.xlabel("Companies", color="white", fontsize=12)
plt.ylabel("Values", color="white", fontsize=12)

# Customize ticks for better visibility
ax.tick_params(axis='x', colors='white')
ax.tick_params(axis='y', colors='white')

# Display the chart using fig1.show()
plt.tight_layout()
fig1.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the devices and their respective NPU scores
devices = ['iPhone 13 Pro Max', 'iPhone 13 Pro', 'iPhone 13 mini',
           'iPad Pro (12.9-inch 5th generation)', 'iPad Pro (11-inch 3rd generation)', 'iPhone 13']
npu_scores = [2676, 2665, 2660, 2657, 2656, 2605]

# Sort the NPU scores and corresponding devices in descending order
sorted_indices = np.argsort(npu_scores)[::-1]
sorted_devices = np.array(devices)[sorted_indices]
sorted_npu_scores = np.array(npu_scores)[sorted_indices]

# Calculate the cumulative sum of NPU scores and the cumulative percentage
cumulative_scores = np.cumsum(sorted_npu_scores)
total_score = cumulative_scores[-1]
cumulative_percentage = cumulative_scores / total_score * 100

# Create a figure and axis
fig6, ax1 = plt.subplots(figsize=(10, 6))

# Plot the bar chart
ax1.bar(sorted_devices, sorted_npu_scores, color='skyblue')
ax1.set_xlabel('Device Name', fontsize=14, color='white')
ax1.set_ylabel('NPU Score', fontsize=14, color='white')
ax1.set_title('Top NPU Scores by Device', fontsize=16, color='white')
ax1.tick_params(axis='x', rotation=45, labelcolor='white')  # Rotate x-axis labels for better readability
ax1.tick_params(axis='y', labelcolor='white')
ax1.set_facecolor('black')  # Set background to black

# Create a second y-axis to plot the cumulative percentage line
ax2 = ax1.twinx()
ax2.plot(sorted_devices, cumulative_percentage, color='red', marker='o', linestyle='-', linewidth=2)
ax2.set_ylabel('Cumulative Percentage (%)', fontsize=14, color='white')
ax2.tick_params(axis='y', labelcolor='white')

# Set the background to black for the second axis as well
ax2.set_facecolor('black')

# Show the chart using fig6.show()
plt.tight_layout()
fig6.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the devices, NPU and GPU scores
devices = ['iPhone 13 Pro Max', 'iPhone 13 Pro', 'iPhone 13 mini',
           'iPad Pro (12.9-inch 5th generation)', 'iPad Pro (11-inch 3rd generation)', 'iPhone 13']
npu_scores = [2676, 2665, 2660, 2657, 2656, 2605]
gpu_scores = [2273, 2261, 2240, 2824, 2842, 2220]

# Create a figure and axis
fig7, ax = plt.subplots(figsize=(10, 6))

# Plotting the stacked bar chart
ax.bar(devices, npu_scores, color='skyblue', label='NPU Score', edgecolor='black')
ax.bar(devices, gpu_scores, bottom=npu_scores, color='orange', label='GPU Score', edgecolor='black')

# Labeling the chart
ax.set_xlabel('Device', fontsize=14, color='white')
ax.set_ylabel('Scores', fontsize=14, color='white')
ax.set_title('NPU and GPU Scores by Device', fontsize=16, color='white')

# Set background color to black and axis labels to white
ax.set_facecolor('black')
ax.tick_params(axis='x', colors='white')  # Change x-axis tick color to white
ax.tick_params(axis='y', colors='white')  # Change y-axis tick color to white

# Add a legend to differentiate the bars and position it outside the plot area
ax.legend(loc='upper left', fontsize=12, bbox_to_anchor=(1.05, 1), borderaxespad=0.)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', color='white')

# Tight layout for better spacing
plt.tight_layout()

# Show the plot using fig7.show()
fig7.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for the devices, CPU scores, and Clock speeds (in GHz for clock speed)
devices = ['iPhone 13 Pro Max', 'iPhone 13 Pro', 'iPhone 13 mini',
           'iPad Pro (12.9-inch 5th generation)', 'iPad Pro (11-inch 3rd generation)', 'iPhone 13']
cpu_scores = [2676, 2665, 2660, 2657, 2656, 2605]
clock_speeds = [3.23, 3.23, 3.10, 3.20, 3.20, 2.65]  # Clock speeds in GHz

# Convert device names to a numpy array for consistent plotting
device_indices = np.arange(len(devices))

# Create a figure and axis
fig8, ax = plt.subplots(figsize=(10, 6))

# Plot the data points for CPU Scores and Clock Speeds
ax.scatter(device_indices, cpu_scores, color='skyblue', label='CPU Score', s=100, edgecolor='black', zorder=5)
ax.scatter(device_indices, np.array(clock_speeds) * 1000, color='orange', label='Clock Speed (MHz)', s=100, edgecolor='black', zorder=5)

# Labeling the chart
ax.set_xlabel('Device', fontsize=14, color='white')
ax.set_ylabel('Scores / Clock Speed (in MHz)', fontsize=14, color='white')
ax.set_title('Comparison of CPU Scores and Clock Speeds by Device', fontsize=16, color='white')

# Set x-axis labels and make them more readable
ax.set_xticks(device_indices)
ax.set_xticklabels(devices, rotation=45, ha='right', color='white')

# Set background color to black and axis labels to white
ax.set_facecolor('black')
ax.tick_params(axis='x', colors='white')  # Change x-axis tick color to white
ax.tick_params(axis='y', colors='white')  # Change y-axis tick color to white

# Add a legend and position it outside the graph
ax.legend(loc='upper left', fontsize=12, bbox_to_anchor=(1.05, 1), borderaxespad=0.)

# Tight layout for better spacing
plt.tight_layout()

# Show the plot using fig8.show()
fig8.show()
```
